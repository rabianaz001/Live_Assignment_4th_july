```python
#for performing visualization
import seaborn as sns
import matplotlib.pyplot as plt
import warnings as wr
wr.filterwarnings('ignore')
```

```python
plt.figure(figsize=(10,6))
sns.histplot(df["AGE"].dropna(), kde=True, color="blue")
plt.title("Distribution of Smoker's Age")
plt.xlabel("Age")
plt.ylabel("SMOKING")
plt.show()
```

```python
alcohol_consumer_counts = df['ALCOHOL_CONSUMING'].value_counts()

plt.figure(figsize=(8,6))
plt.bar(alcohol_consumer_counts.index,alcohol_consumer_counts ,color = 'pink')
plt.xlabel('ALCOHOL_CONSUMING')
plt.ylabel('Count')
plt.title("Count plot of ALCOHOL_CONSUMING")
plt.show()
```

```python
df[ 'LUNG_CANCER'].value_counts().plot.pie(autopct="%1.1f%%",colors=["lightcoral","lightgreen"],explode=(0,0.1),wedgeprops=dict(width=0.5),startangle=90)
plt.title("How many have lung_cancer")
```

```python
cat_col=[col for col in df.columns if df[col].dtypes=="O"]
num_col=[col for col in df.columns if df[col].dtypes!="O"]
```

```python
plt.figure(figsize=(20,15))
plotnumber=1

for col in num_col:
  if plotnumber <= 15:
    ax = plt.subplot(3,5,plotnumber)
    sns.distplot(df[col])
    plt.xlabel(col)
  plotnumber +=1

plt.tight_layout()
plt.show()
```

```python
sns.set_palette("Pastel1")
plt.figure(figsize =(10,8))
sns.pairplot(data = df)
plt.suptitle('Pairplot for Data Frame')
plt.show()
```

```python
plt.figure(figsize=(16,14))
sns.heatmap(df.corr(),annot =True ,fmt = '2f',cmap = 'Pastel2', linewidths= 2)
plt.title("Correlation Heatmap")
plt.show()

#value close to +1 indicates strong positive correalation,-1 indicates a strong negative correlation and 0 indicates suggests no linear correlation.
```
