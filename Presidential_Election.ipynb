```python
# Aggregate the 'will_vote' column based on the 'candidate'
# First, replace 'not sure', 'struggle to answer', 'will not take part' with 0
df['will_vote'] = df['will_vote'].replace({
    'not sure': 0,
    'struggle to answer': 0,
    'will not take part': 0,
    'definitely no': 0,
    'likely no': 0
}).fillna(0).astype(float)

# Now, we sum the votes for each candidate
candidate_votes = df.groupby('candidate')['will_vote'].sum()

# Plotting the pie chart
plt.figure(figsize=(10, 8))
plt.pie(candidate_votes, labels=candidate_votes.index, autopct='%1.1f%%', startangle=140)
plt.title('Vote Distribution by Candidate')
plt.show()
```

```python
voters =pd.crosstab(df["candidate"], df["will_vote"])
voters.plot(kind="bar", stacked=True,figsize=(6,6))
sns.color_palette("bright")
plt.xlabel("candidate")
plt.ylabel("count")
plt.title("Candidate will vote")
plt.show()
```
